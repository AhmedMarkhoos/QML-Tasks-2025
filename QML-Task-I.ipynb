```python
import numpy as np
import pennylane as qml
```

```python
dev = qml.device("default.qubit", wires=5)

@qml.qnode(dev)
def simple_qc():
    for i in range(5): qml.Hadamard(i)
    for i in range(1,5): qml.CNOT([i-1,i])
    qml.SWAP([0,4])
    qml.RX(np.pi/2,0)
    return qml.state()

qml.draw_mpl(simple_qc)()
```

```python
dev = qml.device("default.qubit", wires=4)

@qml.qnode(dev)
def simple_qc():
    qml.Hadamard(0)
    qml.RX(np.pi/3,1)
    qml.Hadamard(2)
    qml.Hadamard(3)
    qml.SWAP([0,1])
    qml.SWAP([2,3])
    return qml.state()

qml.draw_mpl(simple_qc)()
```
